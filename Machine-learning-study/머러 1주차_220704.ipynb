## argsort()
##### 행렬 정렬 후, 그에 대한 인덱스 반환

```python
name_array=np.array(['John','Mike','Sarah','kate','Samuel'])
score_array= np.array([78,95,84,98,88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:',sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:',name_array[sort_indices_asc])
```

## reset_index(drop=True) p.61

인덱스를 설정한 후에 drop=True를 통해 업데이트 해주면, 기존의 인덱스는 삭제됨.

- 참고
https://www.delftstack.com/ko/api/python-pandas/pandas-dataframe-dataframe.reset_index-function/

```python
import pandas as pd
import numpy as np

index = pd.MultiIndex.from_tuples([(1, 'Sarah'),
                                   (1, 'Peter'),
                                   (2, 'Harry'),
                                   (2, 'Monika')],
                                  names=['class', 'name'])
columns = pd.MultiIndex.from_tuples([('Performance', 'max'),
                                     ('Grade', 'type')])
dataframe = pd.DataFrame([('Good', 'A'),
                   ( 'Best', 'A+'),
                   ( 'Bad', 'C'),
                   (np.nan, 'F')],
                  index=index,
                  columns=columns)            
dataframe

## 여기서 인덱스는 class와 name
```

```python
dataframe1 = dataframe.reset_index(drop= True)
dataframe1
#drop=True를 한 결과, 기존의 인덱스는 컬럼으로 다시 들어가지 않고 삭제됨.
```

## iloc vs loc
#### iloc -> 위치 기반 ex) df [행 위치 , 컬럼 위치 ] (숫자)

#### loc -> 명칭 기반 ex) df['index명','컬럼명'] or df['index 번호',' 컬럼명 '] -> 인덱싱의 경우, 종료값 범위까지 포함

## boolen을 사용한 특정 데이터 추출

## 정렬, agg 함수, groupby 

- sort_values

```python
titanic_sorted= titanic_df.sort_values(by=['Name']) #이름을 오름차순으로 정렬 ( default 값 오름차순(ascending=True) )
titanic_sorted.head(3)
```

```python
#여러개의 칼럼 기준으로 정렬
titanic_sorted= titanic_df.sort_values(by=['Name','Pclass'],ascending=False) #pclass와 name 내림차순 정렬
titanic_sorted.head(3)
```

- agg
##### 특정 칼럼에 함수 적용

```python
titanic_df[['Age','Fare']].mean()
```

- groupby

```python
titanic_groupby= titanic_df.groupby('Pclass').count()
titanic_groupby
```

```python
titanic_groupby= titanic_df.groupby('Pclass')[['PassengerId','Survived']].count()
titanic_groupby
```

```python
titanic_groupby= titanic_df.groupby('Pclass')['Age'].agg([max,min])
titanic_groupby
```

```python
# 컬럼마다 각 다른 함수 적용
agg_format={'Age':'max','SibSp':'sum','Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```
